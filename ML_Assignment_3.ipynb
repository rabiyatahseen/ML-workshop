```python
#1.Basic Line Plot
import matplotlib.pyplot as plt
x = list(range(-10, 11))
y = [i**2 for i in x]
plt.plot(x, y)
plt.xlabel("X Values")
plt.ylabel("Squared Values")
plt.title("Line Plot of y = x^2")
plt.show()
```

```python
#2.Bar Plot
import matplotlib.pyplot as plt
students = ['Sai', 'Vamsi', 'Divij', 'Sritan', 'Sita']
marks = [85, 90, 78, 92, 88]
plt.bar(students, marks)
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Marks Scored by Students")
plt.show()
```

```python
#3. Scatter Plot
import matplotlib.pyplot as plt
import numpy as np
x = np.random.randint(0, 50, 20)
y = np.random.randint(0, 50, 20)
plt.scatter(x, y, c='red', s=np.random.randint(100, 500, 20), alpha=0.7, label='Data Points')
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Scatter Plot of Random Data")
plt.legend()
plt.show()
```

```python
# 4. Histogram
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(100)
plt.hist(data, bins=20)
plt.grid(True)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram of Random Normal Data")
plt.show()
```

```python
#5.Pie Chart
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]
percentages = [(value / sum(values)) * 100 for value in values]
colors = ['red', 'blue', 'green', 'orange', 'purple']
plt.pie(values, labels=categories, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title("Percentage Distribution of Categories")
plt.show()
```

```python
#6. Box Plot (Seaborn)
import seaborn as sns
import matplotlib.pyplot as plt
marks = [78, 85, 92, 88, 76, 95, 89, 84, 91, 73]
sns.boxplot(x=marks)
plt.xlabel("Marks")
plt.title("Box Plot of Student Marks")
plt.show()
```

```python
#7. Heatmap (Seaborn)
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
matrix = np.random.randint(1, 11, size=(5, 5))
sns.heatmap(matrix,annot=True, fmt='d')
plt.title('Heatmap of Matrix')
plt.show()
```

```python
#8. Comparison of Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 3]
# Matplotlib Bar Chart
plt.figure(figsize=(8, 4))
plt.bar(x, y)
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Matplotlib Bar Chart')
plt.show()
# Seaborn Bar Chart
plt.figure(figsize=(8, 4))
sns.barplot(x=x, y=y)
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Seaborn Bar Chart')
plt.show()
# Matplotlib line
plt.figure(figsize=(8, 4))
plt.plot(x, y, marker='o')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Matplotlib Line Plot')
plt.show()
# Seaborn line
plt.figure(figsize=(8, 4))
sns.lineplot(x=x, y=y, marker='o')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Seaborn Line Plot')
plt.show()
```

```python
#9. Pair Plot (Seaborn)
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data = {'Age': [25, 30, 22, 28, 35, 27, 21, 32, 29, 24],
        'Height': [170, 180, 165, 178, 185, 172, 168, 182, 175, 169],
        'Weight': [70, 80, 65, 75, 90, 72, 68, 85, 78, 67]}
df = pd.DataFrame(data)
sns.pairplot(df)
plt.suptitle("Pair Plot of Age, Height, and Weight")
plt.show()
```

```python
#10. Save and Style Plots
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 3]
plt.figure(figsize=(8, 4))
sns.lineplot(x=x, y=y, marker='o')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Line Plot with Darkgrid Style')
plt.savefig("line_plot_darkgrid.png")
plt.savefig("line_plot_darkgrid.pdf")
plt.show()
```
